# Quantidade de exemplares emprestados por ano

Extrair o ano da coluna 'data_emprestimo'

```python
# Criar um novo DataFrame 'emprestimos_datetime' que recebe a frequência de emprestimos realizados por cada data
emprestimos_datetime = pd.DataFrame(df['data_emprestimo'].value_counts()).reset_index()

emprestimos_datetime.columns = ['Datetime','Quantidade']

# Converter a coluna 'Datetime' para datetime
emprestimos_datetime['Datetime'] = pd.to_datetime(emprestimos_datetime['Datetime'])

emprestimos_datetime
```

Gerar um gráfico de linha para acompanhar a mudança na quantidade.

```python
emprestimos_por_ano = emprestimos_datetime.groupby(by=emprestimos_datetime.Datetime.dt.year)['Quantidade'].sum()
emprestimos_por_ano.index.name = 'Ano'
```

```python
plt.figure(figsize=(10, 6))
emprestimos_por_ano.plot(kind='line', marker='o')
plt.xlabel('Ano')
plt.ylabel('Quantidade de Exemplares Emprestados')
plt.title('Quantidade de Exemplares Emprestados por Ano')
plt.grid(True)
plt.show()
```

Percebe-se que a partir de 2013 a quantidade de exemplares emprestados vem diminuindo... Conseguiu melhorar um pouco em 2017, mas após isso a queda na quantidade intensificou muito.

- O que aconteceu de 2017 até 2020?

Obs.: Importante destacar que não temos os dados de todos os meses de 2020, apenas do 1º semestre.


# Quantidade de exemplares emprestados por mês ao longo dos anos

Agora vamos analisar mais detalhadamente. Verificar como essa quantidade oscila em cada mês.

```python
emprestimos_por_mes = emprestimos_datetime.groupby(by=emprestimos_datetime.Datetime.dt.month)['Quantidade'].sum()
emprestimos_por_mes.index.name = 'Mês'

dict_meses = {1:'Jan',2:'Fev',3:'Mar',4:'Abr',
                    5:'Mai',6:'Jun',7:'Jul',8:'Ago',
                    9:'Set',10:'Out',11:'Nov', 12:'Dez'}

emprestimos_por_mes.index = emprestimos_por_mes.index.map(dict_meses)
```

```python
plt.figure(figsize=(10, 6))
emprestimos_por_mes.plot(kind='line', marker='o')
plt.xlabel('Meses')
plt.ylabel('Quantidade de Exemplares Emprestados')
plt.title('Quantidade de Exemplares Emprestados por Mês')
plt.grid(True)

plt.show()
```

# Gerenciar os recursos humanos da biblioteca, com base na demanda de trabalho

- Gerenciar a programação de férias dos colaboradores de acordo com os meses de menor demanda;
- Programar atividades que não sejam de atendimento ao usuário para períodos específicos de menor demanda.

```python
plt.figure(figsize=(10, 6))

plt.bar(emprestimos_por_mes.index, emprestimos_por_mes.values)

plt.title('Quantidade Total de Exemplares Emprestados por Mês (2010-2020)')
plt.xlabel('Mês')
plt.ylabel('Quantidade de Exemplares Emprestados')

plt.show()
```

Com base nas mesmas variaveis do gráfico anterior, podemos visualizar melhor agora que, sempre nos meses de férias como **Janeiro, Junho, Julho e Dezembro, a demanda por empréstimo cai.**

Ou seja, o melhor período para os funcionários tirarem **férias** é no período de férias acadêmicas, **Jun/Jul e Jan/Dez**

```python
emprestimos_por_hora = emprestimos_datetime.groupby(by=emprestimos_datetime.Datetime.dt.hour)['Quantidade'].sum()
emprestimos_por_hora.index.name = 'Horário'

emprestimos_por_hora
```

```python
plt.figure(figsize= (10, 6))

plt.bar(emprestimos_por_hora.index, emprestimos_por_hora.values)

plt.xlabel('Horário')
plt.ylabel('Quantidade de Exemplares Emprestados por hora')

plt.show()

### IMPRIR HORA POR HORA
```
